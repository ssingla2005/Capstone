```python
from google.colab import drive
import os
import shutil
from sklearn.model_selection import train_test_split
from ultralytics import YOLO
import time # Added for safety checks

# --- Setup and Mount Drive ---
print("Mounting Google Drive...")
drive.mount('/content/drive', force_remount=True)

# Verify and set paths
BASE_DATA_PATH = '/content/drive/My Drive/dataest_drowsiness/train'
CLASS_NAMES = ['Closed', 'Open', 'Yawn', 'No_yawn']
YOLO_DATA_ROOT = '/content/yolo_drowsiness_data'
TRAIN_DIR = os.path.join(YOLO_DATA_ROOT, 'train')
VAL_DIR = os.path.join(YOLO_DATA_ROOT, 'val')

# --- Create YOLO Dataset Structure and Split Data ---
print("\n--- Structuring Dataset for YOLO (80% Train, 20% Val) ---")
os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(VAL_DIR, exist_ok=True)
for class_name in CLASS_NAMES:
    os.makedirs(os.path.join(TRAIN_DIR, class_name), exist_ok=True)
    os.makedirs(os.path.join(VAL_DIR, class_name), exist_ok=True)

total_train_images = 0
for class_name in CLASS_NAMES:
    class_src_dir = os.path.join(BASE_DATA_PATH, class_name)
    if not os.path.exists(class_src_dir):
      print(f"ðŸ›‘ Warning: Source directory not found: {class_src_dir}. Skipping.")
      continue

    all_files = [os.path.join(class_src_dir, f) for f in os.listdir(class_src_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    train_files, val_files = train_test_split(all_files, test_size=0.2, random_state=42)

    # Copy files
    for f in train_files:
        shutil.copy(f, os.path.join(TRAIN_DIR, class_name))
    for f in val_files:
        shutil.copy(f, os.path.join(VAL_DIR, class_name))

    total_train_images += len(train_files)
    print(f"Class {class_name}: Train={len(train_files)}, Val={len(val_files)}")
print(f"Data preparation complete. Total Training Images: {total_train_images}")


# --- Train the YOLO Classification Model ---
print("\n--- Training YOLOv8n Classification Model ---")

model = YOLO('yolov8n-cls.pt')

NUM_EPOCHS = 50
IMG_SIZE = 145
BATCH_SIZE = 32

print(f"Starting training for {NUM_EPOCHS} epochs...")
results = model.train(
    data=YOLO_DATA_ROOT,
    epochs=NUM_EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    name='drowsiness_classifier',
    project='yolo_runs',
    exist_ok=True,
    cache=False,
    verbose=True
)

print("\n--- TRAINING COMPLETE ---")
```

```python


import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
! pip install ultralytics
from ultralytics import YOLO # Ensure this import is successful now!

# --- Configuration ---
RUNS_DIR = 'yolo_runs'
PROJECT_NAME = 'drowsiness_classifier'
VAL_DIR = '/content/yolo_drowsiness_data/val'
CLASS_NAMES = ['Closed', 'Open', 'Yawn', 'No_yawn']

# Construct the path to the best weights
BEST_WEIGHTS_PATH = os.path.join(RUNS_DIR, PROJECT_NAME, 'weights/best.pt')

if not os.path.exists(BEST_WEIGHTS_PATH):
    print("ðŸ›‘ Error: Best weights not found. Did Cell 2 (Training) complete successfully?")
else:
    # --- Load the best model and proceed ---
    best_model = YOLO(BEST_WEIGHTS_PATH)

    # --- 1. Deep Dive into Evaluation Metrics ---
    print("\n## ðŸ“Š 1. Deep Dive into Evaluation Metrics")
    metrics = best_model.val()

    results_path = os.path.join(RUNS_DIR, PROJECT_NAME)
    print(f"âœ… Validation Top-1 Accuracy: {metrics.top1:.4f}")
    print(f"Check the Confusion Matrix at: {results_path}/confusion_matrix.png")
    #
    # --- 2. Hyperparameter Tuning (Evolution) - SKIPPED ---
    print("\n## ðŸ“ˆ 2. Hyperparameter Tuning (Evolution) - SKIPPED DUE TO PREVIOUS ERRORS.")

    # --- 3. Real-Time Deployment Simulation ---
    print("\n## ðŸ› ï¸ 3. Real-Time Deployment Simulation")

    # Drowsiness Logic Parameters
    DROWSINESS_THRESHOLD = 0.85
    CLOSED_EYE_COUNT = 0
    YAWN_COUNT = 0
    MAX_CLOSED_FRAMES = 10
    ALARM_TRIGGERED = False
    SIMULATION_FRAMES = 15

    # 1. Get a sequence of random validation images for simulation
    all_val_files = []
    for c in CLASS_NAMES:
        class_path = os.path.join(VAL_DIR, c)
        if os.path.exists(class_path):
            all_val_files.extend([os.path.join(class_path, f) for f in os.listdir(class_path)])

    simulation_sequence = random.sample(all_val_files, min(SIMULATION_FRAMES, len(all_val_files)))

    # 2. Simulate frame-by-frame processing
    print(f"Simulating real-time feed with {len(simulation_sequence)} frames...")
    plt.figure(figsize=(15, 6))
    fig_count = 1

    for img_path in simulation_sequence:

        # Run prediction
        results = best_model.predict(source=img_path, verbose=False)

        # Extract prediction
        probs = results[0].probs
        predicted_class_name = best_model.names[probs.top1]
        confidence = probs.top1conf.item()

        # --- Drowsiness Alert Logic ---
        if predicted_class_name == 'Closed' and confidence > DROWSINESS_THRESHOLD:
            CLOSED_EYE_COUNT += 1
            YAWN_COUNT = 0
            status_color = 'red'
        elif predicted_class_name == 'Yawn' and confidence > DROWSINESS_THRESHOLD:
            YAWN_COUNT += 1
            CLOSED_EYE_COUNT = 0
            status_color = 'red'
        else:
            CLOSED_EYE_COUNT = max(0, CLOSED_EYE_COUNT - 1)
            YAWN_COUNT = max(0, YAWN_COUNT - 1)
            status_color = 'green'

        # Check for alarm trigger
        if (CLOSED_EYE_COUNT >= MAX_CLOSED_FRAMES or YAWN_COUNT >= MAX_CLOSED_FRAMES) and not ALARM_TRIGGERED:
            ALARM_TRIGGERED = True
            print(f"\nðŸš¨ðŸš¨ DROWSINESS ALARM ACTIVATED at frame {fig_count}! ðŸš¨ðŸš¨")


        # 3. Visualize the result for this frame
        img = cv2.imread(img_path)
        if img is None: continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        ax = plt.subplot(3, 5, fig_count)
        plt.imshow(img)
        plt.title(f"P: {predicted_class_name}\nConf: {confidence:.2f}\nFrames: {max(CLOSED_EYE_COUNT, YAWN_COUNT)}", fontsize=8, color=status_color)
        plt.axis("off")
        fig_count += 1

    plt.suptitle("Real-Time Simulation of Drowsiness Detection Logic", fontsize=16)
    plt.tight_layout()
    plt.show()

    print(f"\nSimulation End: Final Alarm Status: {'TRIGGERED' if ALARM_TRIGGERED else 'IDLE'}")
```
